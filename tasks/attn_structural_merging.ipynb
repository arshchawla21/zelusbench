# ZelusBench: Structural Attention — MergingPins leaf_bias=0.25 while randomizing everything else.Controls graph topology from linear chains to bushy/diamond structures.- 15 scenarios

```python
import kaggle_benchmarks as kbenchimport numpy as npimport randomimport timefrom zelusbench.scenarios.config import ScenarioConfig, QueryTypefrom zelusbench.scenarios.generator import ScenarioGeneratorfrom zelusbench.evaluation.scorer import score_query, score_responseLEAF_BIAS = 0.25SEEDS = 15print(f"ZelusBench Structural Attention — Merging (leaf_bias={LEAF_BIAS})")
```

```python
@kbench.task(name="zelusbench_attn_structural_merging")def zelusbench_attn_structural_merging(llm) -> tuple[float, float]:    _start = time.time()    print(f"Running {SEEDS} scenarios...")    print("=" * 60)    all_scores = []    for i in range(SEEDS):        bg_rng = random.Random(i * 7919)        cfg = ScenarioConfig.randomize_except(bg_rng, pinned={            "leaf_bias": LEAF_BIAS,            "num_queries": 3,            "seed": i,        })        scenario = ScenarioGenerator(cfg).generate(f"structural_{i}")        response = llm.prompt(scenario.prompt)        scores = score_response(response, scenario)        all_scores.extend(scores)        avg = float(np.mean([s.score for s in scores]))        bg = f"dim={cfg.dim} pts={cfg.num_points} tp={cfg.transform_prob}"        print(f"  [{i+1}/{SEEDS}] avg={avg:.2f} | {bg}")    overall = float(np.mean([s.score for s in all_scores]))    std = float(np.std([s.score for s in all_scores]))    print(f"\nOverall: {overall:.3f} +/- {std:.3f} | {len(all_scores)} queries | {time.time()-_start:.1f}s")    kbench.assertions.assert_true(overall >= 0, expectation=f"Valid answers (got {overall:.3f})")    return overall, stdzelusbench_attn_structural_merging.run(llm=kbench.llm)
```
